```python
df = df.drop(columns=['Timestamp','Participant_ID'])
df.info()
```

# Add 'Age category' feature 

```python
a  = df['Age']
print(len(a))
```

```python

def age_category(age):
    if age <=25:
        return 'Youth'
    elif 25< age <=40:
        return 'Middle'
    else:
        return 'Old'

df['Age category'] = df['Age'].apply(age_category)    # ! .apply() orqali yangi feature yasaldi
```

```python
df['Age category'].value_counts()
```

# Fill missing value

```python
df.isnull().sum() 
```

```python
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df.isnull().sum()
```

```python
from sklearn.preprocessing import LabelEncoder        # Encoding
for col in  df.columns :
  if df[col].dtypes == 'object':
    df[col] = LabelEncoder().fit_transform(df[col])
df.head(20)
```

```python
from sklearn.preprocessing import StandardScaler          # Scaling
# df= pd.DataFrame(StandardScaler().fit_transform(df), columns=df.columns)      # !  [[]]  changes 1D to 2Dimension
df[['Age', 'Nationality']]= StandardScaler().fit_transform(df[['Age','Nationality']])
df.head(20)

```
