# Add 'Age category' feature 

```python
a  = df['Age']
print(len(a))
```

```python

def age_category(age):
    if age <=25:
        return 'Youth'
    elif 25< age <=40:
        return 'Middle'
    else:
        return 'Old'

df['Age category'] = df['Age'].apply(age_category)    # ! .apply() orqali yangi feature yasaldi
```

```python
df['Age category'].value_counts()
```

```python
from sklearn.preprocessing import LabelEncoder
for col in  df.columns :
  if df[col].dtypes == 'object':
    df[col] = LabelEncoder().fit_transform(df[col])
df.head()
```

```python
from sklearn.preprocessing import StandardScaler
df['Age'] = StandardScaler().fit_transform()
```
