# Encoding df['species'] with map()

```python
df['species']  = df['species'].map({
    'Iris-setosa' : 1,
    'Iris-versicolor': 2,
    'Iris-virginica' : 3
}
)
```

```python
df.info()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score, mean_squared_error
```

```python
x = df.drop(columns = ['species'])
y = df['species']
x_train ,x_temp,y_train,y_temp=train_test_split(x,y,test_size=0.2,random_state=42)
x_test,x_val,y_test,y_val=train_test_split(x_temp,y_temp,test_size=0.5,random_state=42)
```

# Linear model

```python
linear_model = LinearRegression().fit(x_train,y_train)
linear_model
```

```python
y_pred = linear_model.predict(x_test)
linear_r2 = r2_score(y_test,y_pred)
linear_mse = mean_squared_error(y_test,y_pred)
```

# DecisionTree model

```python
dtree_model = DecisionTreeRegressor().fit(x_train,y_train)
dtree_model
```

```python
y_pred = dtree_model.predict(x_test)
dtree_r2 = r2_score(y_test,y_pred)
dtree_mse = mean_squared_error(y_test,y_pred)
```

# Random Forest model

```python
rforest_model = RandomForestRegressor().fit(x_test,y_test)
rforest_model
```

```python
y_pred = rforest_model.predict(x_test)
rforest_r2 = r2_score(y_test,y_pred)
rforest_mse = mean_squared_error(y_test,y_pred)
```

```python
from tabulate import tabulate
headers = ['Model Name', 'r2_score' , "mean_squared_error"]
table = [['linear_model', 'linear_r2', "linear_mse"],
         ['dtree_model', 'dtree_r2', 'dtree_mse'],
         ['rforest_model', 'rforest_r2', 'rforest_mse']
]

print(tabulate(table, headers, tablefmt='grid'))
```
